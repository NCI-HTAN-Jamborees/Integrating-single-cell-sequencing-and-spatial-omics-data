# 313 integration

```python
sc_adata = sc.read_h5ad(os.path.join(scRNA_folder,'HTAPP-313-SMP-932_fresh_channel1.h5ad'))
```

```python
sc_adata.obs
```

```python
sc_adata.var
```

```python
sp_adata = sc.read_h5ad(os.path.join(merfish_folder, 'HTAPP-313_centroids.h5ad'))
```

```python
sp_adata.obs
```

```python
xs = sp_adata.obs.x_centroid.values
ys = sp_adata.obs.y_centroid.values
plt.axis('off')
plt.scatter(xs, ys, s=.7);
plt.gca().invert_yaxis()
```

```python
sc.pp.normalize_total(sc_adata)
```

```python
sc.tl.rank_genes_groups(sc_adata, groupby="cell_groups", use_raw=False)
markers_df = pd.DataFrame(sc_adata.uns["rank_genes_groups"]["names"]).iloc[0:100, :]
markers = list(np.unique(markers_df.melt().value.values))
len(markers)
```

```python
tg.pp_adatas(sc_adata, sp_adata, genes=markers)
```

```python
ad_map = tg.map_cells_to_space(adata_sc=sc_adata,
                               adata_sp=sp_adata,
                               # mode="clusters",
                               # cluster_label='cell_groups',
                               # density_prior='rna_count_based',
                               # num_epochs=500,
                               device='cpu',)
```

```python
ad_map
```

## mapping single cells into space

```python
tg.project_cell_annotations(ad_map, sp_adata, annotation="cell_groups")
annotation_list = list(pd.unique(sc_adata.obs['cell_groups']))
tg.plot_cell_annotation_sc(sp_adata, annotation_list,x='x_centroid', y='y_centroid', spot_size=50, perc=0.02)
```

```python
tg.plot_training_scores(ad_map, bins=10, alpha=.5)
```

```python
ad_ge = tg.project_genes(adata_map=ad_map, adata_sc=sc_adata)
```

```python
ad_ge
```

```python
ad_map.uns['train_genes_df'].sort_values(by='train_score').head(25)
```

```python
genes = ['cd63', 'ifitm3',  'cst3']
ad_map.uns['train_genes_df'].loc[genes]
```

```python
tg.plot_genes_sc(genes, adata_measured=sp_adata, adata_predicted=ad_ge, spot_size=50,scale_factor=0.1, perc=0.02, return_figure=False)
```

```python
ad_ge.write('/sbgenomics/output-files/313_ad_ge.h5ad')
```

```python
scRNA_folder = '/sbgenomics/project-files/filtered_scRNA-seq'
merfish_folder = '/sbgenomics/project-files/processed_MERFISH'
finished_folder = '/sbgenomics/project-files/tangram_integration'
output_folder = '/sbgenomics/output-files'
celltype_folder = os.path.join(scRNA_folder, 'CellType')
```

```python
for file in os.listdir(scRNA_folder)[-2:]:
    print (file)
```

```python
for file in os.listdir(scRNA_folder)[-2:]:
    if file.endswith('.h5ad'):
        case_id = 'HTAPP-' + file.split('-')[1]
        sc_full_path = os.path.join(scRNA_folder, file)
        sp_full_path = os.path.join(merfish_folder, case_id + '_centroids.h5ad')
        # print (sc_full_path)
        # print (sp_full_path)
        # associated_metadata = os.path.join(celltype_folder, case_id + '_cell_type.csv')

        # output files
        map_filename = os.path.join(output_folder, os.path.splitext(file)[0] + '_ad_map.h5ad')
        ge_filename = os.path.join(output_folder, os.path.splitext(file)[0] + '_ad_ge.h5ad' )
        finished_file = os.path.join(finished_folder, os.path.splitext(file)[0] + '_ad_map.h5ad')
        # print (finished_file)

        if os.path.exists(map_filename):
            print (map_filename + ' already integrated')
        elif os.path.exists(finished_file):
            print (finished_file + ' already integrated')
        else:
            print (file + ' to be integrated')
            # load data and metadata
            sc_adata = sc.read_h5ad(sc_full_path)
            sp_adata = sc.read_h5ad(sp_full_path)
            # # metadata = pd.read_csv(associated_metadata, index_col=0, header=0)
            # # filtered_metadata = metadata[metadata['HTAPP_id'] == os.path.splitext(file)[0]]
    
            # generate markers
            sc_adata.var_names_make_unique()
            sc.pp.normalize_total(sc_adata)
            sc.tl.rank_genes_groups(sc_adata, groupby="cell_groups", use_raw=False)
            markers_df = pd.DataFrame(sc_adata.uns["rank_genes_groups"]["names"]).iloc[0:100, :]
            markers = list(np.unique(markers_df.melt().value.values))
            # len(markers)
    
            tg.pp_adatas(sc_adata, sp_adata, genes=markers)
    
            # map
            ad_map = tg.map_cells_to_space(adata_sc=sc_adata,
                                   adata_sp=sp_adata,
                                   # mode="clusters",
                                   # cluster_label='cell_groups',
                                   # density_prior='rna_count_based',
                                   # num_epochs=500,
                                   device='cpu',)
    
            ad_ge = tg.project_genes(adata_map=ad_map, adata_sc=sc_adata)
            
            # # light preprocessing
            # raw_adata.obs['cell_groups'] = filtered_metadata['Cell Subset']
            # adata = raw_adata[~raw_adata.obs['cell_groups'].isna()]
            # adata.var_names_make_unique()
            # sc.pp.normalize_total(adata)
    
            # print (filename)
            # filtered_filename = os.path.join(output_folder, filename)
            # print (filtered_filename)
            ad_map.write(map_filename)
            ad_ge.write(ge_filename)
            
```
