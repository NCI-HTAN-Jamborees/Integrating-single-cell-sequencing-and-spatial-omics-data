# initial QC

```python
sc.pl.highest_expr_genes(adata, n_top=20, )
```

```python
sc.pp.filter_cells(adata, min_genes=200)
sc.pp.filter_genes(adata, min_cells=3)
```

```python
adata.var['mt'] = adata.var_names.str.startswith('MT-')  # annotate the group of mitochondrial genes as 'mt'
sc.pp.calculate_qc_metrics(adata, qc_vars=['mt'], percent_top=None, log1p=False, inplace=True)
```

```python
sc.pl.violin(adata, ['n_genes_by_counts', 'total_counts', 'pct_counts_mt'],
             jitter=0.4, multi_panel=True)
```

# shrink h5ads

```python
scRNA_folder = '/sbgenomics/project-files/HTAN-HTAPP/scRNA-seq'
filtered_scRNA_folder = '/sbgenomics/output-files'
celltype_folder = os.path.join(scRNA_folder, 'CellType')
```

```python
for file in os.listdir('/sbgenomics/project-files/HTAN-HTAPP/scRNA-seq'):
    if file.endswith('.h5ad'):
        case_id = 'HTA1_' + file.split('-')[1]
        full_path = os.path.join(scRNA_folder, file)
        print (full_path)
        associated_metadata = os.path.join(celltype_folder, case_id + '_cell_type.csv')

        # load data and metadata
        raw_adata = sc.read_h5ad(full_path)
        metadata = pd.read_csv(associated_metadata, index_col=0, header=0)
        filtered_metadata = metadata[metadata['HTAPP_id'] == os.path.splitext(file)[0]]

        # light preprocessing
        raw_adata.obs['cell_groups'] = filtered_metadata['Cell Subset']
        adata = raw_adata[~raw_adata.obs['cell_groups'].isna()]
        adata.var_names_make_unique()
        sc.pp.normalize_total(adata)

        # output files
        filtered_filename = os.path.join(filtered_scRNA_folder, file)
        adata.write(filtered_filename)
```
